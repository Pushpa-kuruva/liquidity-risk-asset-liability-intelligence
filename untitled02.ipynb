```python
import pandas as pd
df_loan=pd.read_csv("C:/Users/admin/Downloads/datasets/loan_portfolio_dataset.csv")
```

```python
df_loan
```

```python
df_loan.head()
```

```python
df_loan = df_loan.dropna(subset=['loan_id'])
```

```python
df_loan
```

```python
num_cols = df_loan.select_dtypes(include=['int64', 'float64']).columns
```

```python
num_cols
```

```python
df_loan[num_cols] = df_loan[num_cols].apply(lambda x: x.fillna(x.median()))
```

```python
df_loan[num_cols]
```

```python
cat_cols = df_loan.select_dtypes(include=['str']).columns

df_loan[cat_cols] = df_loan[cat_cols].apply(lambda x: x.str.lower().str.strip())
```

```python
df_loan[cat_cols]
```

```python
df_loan['maturity_bucket'] = df_loan['repricing_frequency'].str.lower().str.strip()
```

```python
df_loan['maturity_bucket']
```

```python

df_loan_financial_cols = df_loan[['sanction_amount', 'interest_rate', 'outstanding_principal']]
```
