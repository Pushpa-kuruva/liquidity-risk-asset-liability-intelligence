```python
df_loan['tenure_months'] = pd.to_numeric(df_loan['tenure_months'])
```

```python
df_loan['tenure_months'] 
```

```python
import numpy as np

conditions = [
    (df_loan['tenure_months'] >= 0) & (df_loan['tenure_months'] <= 1),
    (df_loan['tenure_months'] > 1) & (df_loan['tenure_months'] <= 3),
    (df_loan['tenure_months'] > 3) & (df_loan['tenure_months'] <= 6),
    (df_loan['tenure_months'] > 6) & (df_loan['tenure_months'] <= 12),
    (df_loan['tenure_months'] > 12)
]

choices = [
    '0-30 Days',
    '30-90 Days',
    '90-180 Days',
    '6M-1Y',
    '1Y+'
]
df_loan['alm_bucket'] = np.select(conditions, choices, default='unknown')
```

```python
df_loan['alm_bucket']
```

```python
df_loan['is_floating'] = df_loan['repricing_frequency'] != 'Fixed'
```
