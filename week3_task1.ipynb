```python
def color_liquidity(val):
    if val < 0:
        return 'background-color: red; color: white'
    elif val < 50:
        return 'background-color: yellow; color: black'
    else:
        return 'background-color: green; color: white'
```

```python
def business_interpretation(val):
    if val < 0:
        return "Immediate risk | Need funding"
    else:
        return "Healthy"

df_macro['business_interpretation'] = df_macro['liquidity_index'].apply(business_interpretation)

df_macro[['date', 'liquidity_index', 'business_interpretation']].head()
```

```python
def highlight_risk(val):
    if val < 0:
        return 'background-color: red; color: white'
    else:
        return ''

df_macro.head(5).style.map(highlight_risk, subset=['liquidity_index'])
```

```python
def liquidity_gap_rule(row):
    if row['outflow_type'] > row['inflow_type']:
        return "Liquidity gap forming"
    else:
        return "No gap"
```

```python
df_cash['liquidity_gap_status'] = df_cash.apply(liquidity_gap_rule, axis=1)

df_cash[['inflow_type', 'outflow_type', 'liquidity_gap_status']]
```
