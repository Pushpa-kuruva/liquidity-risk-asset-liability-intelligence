```python
def assign_bucket(days):
    if days <= 30:
        return '0-30 days'
    elif days <= 90:
        return '30-90 days'
    elif days <= 180:
        return '90-180 days'
    elif days <= 365:
        return '180-365 days'
    else:
        return '1 year+'
```

```python
import pandas as pd
df_borr=pd.read_csv("C:/Users/admin/Downloads/datasets/borrowing_funding_dataset.csv")
```

```python
df_borr['bucket']= df_borr['maturity_bucket']
```

```python
liability_exposure = (df_borr.groupby('maturity_bucket')['borrowing_amount'].sum())
```

```python
liability_exposure
```

```python
asset_exposure = (df_loan.groupby('bucket')['outstanding_principal'].sum())
```

```python
asset_exposure
```

```python
ALM_Gap = asset_exposure - liability_exposure
```

```python
ALM_Gap
```

```python
def interpret_gap(gap):
    if gap > 0:
        return "Liquidity Surplus"
    elif gap < 0:
        return "Liquidity Deficit"
    else:
        return "Balanced"
```

```python
Interpretation = ALM_Gap.apply(interpret_gap)
```

```python
Interpretation 
```

```python
Cumulative_Gap = ALM_Gap.cumsum()
```

```python
Cumulative_Gap
```
