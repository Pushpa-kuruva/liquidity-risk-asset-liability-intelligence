```python
def convert_to_months(x):
    x = str(x).strip().upper()
    
    if "M" in x:
        return int(x.replace("M", ""))
    elif "Y" in x:
        return int(x.replace("Y", "")) * 12
    else:
        return None

df_borr["maturity_months"] = df_borr["maturity_bucket"].apply(convert_to_months)
```

```python
df_borr["maturity_months"]
```

```python
alm = df_borr.groupby("maturity_months")["borrowing_amount"].sum().reset_index()
alm = alm.sort_values(by="maturity_months")

alm
```

```python
alm["inflow"] = 0  
alm["outflow"] = df_borr["borrowing_amount"]

alm["gap"] = alm["inflow"] - alm["outflow"]
```

```python
alm["gap"] 
```

```python
def color_gap(val):
    if val < 0:
        return "background-color: red"
    elif abs(val) < 1e-6:   # near zero
        return "background-color: yellow"
    else:
        return "background-color: green"

alm.style.map(color_gap, subset=["gap"])
```

```python
short_term = df_borr[df_borr["maturity_months"] <= 12]["borrowing_amount"].sum()
total = df_borr["borrowing_amount"].sum()

rollover_ratio = short_term / total
```

```python
rollover_ratio
```

```python
alm["rollover_risk_flag"] = "LOW"

if rollover_ratio > 0.6:
    alm["rollover_risk_flag"] = "HIGH"
```

```python
def highlight_rollover(val):
    if val == "HIGH":
        return "background-color: red"
    return ""

alm.style.map(highlight_rollover, subset=["rollover_risk_flag"])
```
