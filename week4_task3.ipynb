```python
import pandas as pd

def convert_to_months(x):
    x = str(x).strip().upper()
    
    if "M" in x:
        return int(x.replace("M", ""))
    elif "Y" in x:
        return int(x.replace("Y", "")) * 12
    else:
        return None

df_borr["maturity_months"] = df_borr["maturity_bucket"].apply(convert_to_months)
```

```python
df_borr["maturity_months"]
```

```python
df_borr["short_term_funding"] = df_borr["maturity_months"] <= 12
```

```python
df_borr["short_term_funding"]
```

```python
total_funding = df_borr["borrowing_amount"].sum()

short_term_funding = df_borr.loc[df_borr["short_term_funding"], "borrowing_amount"].sum()
```

```python
funding_risk_ratio = short_term_funding / total_funding
```
