```python
import pandas as pd

df_loan = df_loan.copy()
numeric_cols = df_loan.select_dtypes(include="number").columns

summary = []

for col in numeric_cols:
    Q1 = df_loan[col].quantile(0.25)
    Q3 = df_loan[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df_loan[(df_loan[col] < lower_bound) | (df_loan[col] > upper_bound)]
    
    summary.append({
        "column": col,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "outlier_count": len(outliers),
        "outlier_%": round(len(outliers) / len(df_loan) * 100, 4)
    })

outlier_summary = pd.DataFrame(summary)

outlier_summary
```
